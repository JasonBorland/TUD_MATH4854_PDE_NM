\therefore y &= Bxe^{-6x}\\
\text{Using $y(2\pi) = 0$} \\
y(2\pi) &= 0 = 2\pi Be^{-12\pi}\\
\therefore B = 0, \ \ \text{As $2\pi e^{-12\pi} > 0$} \\
\end{align}
$$

So only the trivial solution with case 2.


#### Case 3: 
$\lambda > 0, \therefore -\lambda < 0$

Let $\lambda = k^2$, where $k > 0$.
So we have $r = -6 \pm ik$, which yields solution:

$$
\begin{align}
y &= Ae^{(-6+ik)x} + Be^{(-6-ik)x} \\
&= e^{-6x}(Ae^{ikx} + Be^{-ikx}) \\
&= e^{-6x}(Dcos(kx) + Csin(kx)) \\
& \text{Using y(0) = 0} \\
y(0) &= 0 = (1)(Dcos(0) + Csin(0)) \\
\therefore D &= 0 \\
y &= Ce^{-6x}sin(kx) \\
&\text{Using $y(2\pi) = 0$} \\
y(2\pi) &= 0 = Ce^{-12\pi}sin(2\pi k) \\
&\text{We want $C \ne 0$, and we know $e^{-12\pi} \ne 0$} \\
\therefore sin(2\pi k) &= 0 \\
\therefore 2\pi k &= n \pi, \ \text{where $n = 1,2,3, \dots $}\\
\therefore k&= \frac{n}{2} \\
\end{align}
$$

THe eigenfunction is $ y_n = C_nsin(n\pi)$, where $n = 1,2,3, \dots $

The eigenvalue is $\lambda_n = k^2 = \frac{n^2}{4}$
